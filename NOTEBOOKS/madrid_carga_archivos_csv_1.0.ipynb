### CONVERSIÓN DE TIPOS

Pasamos a convertir los tipos de los campos (todos detectados inicialmente como Object) que no sean texto. En el caso de los campos relativos a las condiciones meteorológicas y de suelo aprovecharmos para codificarlas de booleano a número por mapeo.

```python
conv_tipos = {'numero': ['NUM_CALLE','NUM_VICTIMAS'], 'fecha': ['FECHA'], \
              'booleano': ['Granizo','Hielo','Lluvia','Niebla','Seco','Nieve','supf_Mojada','supf_Aceite','supf_Barro', \
                          'supf_Grava_Suelta', 'supf_Hielo', 'supf_Seca_Limpia']}

conv_bool = {'NO': 0, 'SI':1}
  

for tipo, columnas in conv_tipos.items():
      
    if tipo == 'numero':
        
        for col in columnas:
            dfmadrid[col] = pd.to_numeric(dfmadrid[col])
            
    if tipo == 'fecha':
        for col in columnas:
            dfmadrid[col] = pd.to_datetime(dfmadrid[col], dayfirst=True, format="%d/%m/%Y")
            
    if tipo == 'booleano':
        for col in columnas:
            dfmadrid[col] = dfmadrid[col].map(conv_bool)
            
            
dfmadrid.dtypes
```
