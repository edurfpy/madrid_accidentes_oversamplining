### CAMBIO ESQUEMA DEL DATASET.

En los archivos originales cada registro (fila) representaba una víctima. Hay gran número de campos relativos al propio accidente (de campo inicial 'FECHA' al campo 'TIPO_ACCIDENTE', total 21 columnas), y los últimos 5 campos relativos a la víctima. Eso produce que si un accidente tuvo 6 víctimas habrá 6 registros con 21 campos comunes (los relativos al accidente, que son el mismo) y solo son distintos los últimos 5 campos de víctima.

El estudio es sobre accidentes. Es necesario cambiar el esquema del dataset. Dividiremos el dataset original en estas dos partes (camps accidente / campos victima). Del primero eliminaremos registros repetidos y del segundo resumiremos la información de las distintas victimas de cada accidente en campos agregados/resumen de interés para nuestro estudio.

También haremos otras tareas de preprecesamiento necesario (agrupar niveles en atributos categóricos, etc.)

### AGREGAR/RESUMIR DATOS VICTIMAS DE CADA ACCIDENTE.

**PROPUESTA FINAL AGREGACION/RESUMEN DATOS VICTIMAS MISMO ACCIDENTE**

- número vehiculos implicados accidente
- nº hombres y mujeres victimas
- nº heridos leves, graves; nº muertos, nº ilesos
- nº conductores por edad. Se aprovecha para reducir los tramos de edad, que iban de 5 en 5 (alguno menos incluso), sin interés para el estudio y demasiados.

```python
def unir_estad_victimas3(df_in):
    
    num_parte = df_in['NUM_PARTE'].unique()[0]
   

    df_conductor = df_in[df_in['TIPO_PERSONA'] == 'CONDUCTOR']
    num_veh = len(df_conductor)
        
    n_cond_menor = n_cond_18 = n_cond_35 = n_cond_50 = n_cond_desconoc = 0
    if len(df_conductor) != 0:
        edad_conductor = df_conductor['Tramo_Edad'].value_counts()
        N = len(edad_conductor)
        for i in range(N):
            if ((edad_conductor.index[i]== 'DE 0 A 5 AÑOS') | (edad_conductor.index[i]== 'DE 6 A 9 AÑOS') | \
                (edad_conductor.index[i]== 'DE 10 A 14 AÑOS') | (edad_conductor.index[i]== 'DE 15 A 17 AÑOS')):
                
                n_cond_menor = n_cond_menor + edad_conductor.values[i]
                
            elif ((edad_conductor.index[i]== 'DE 18 A 20 AÑOS') | (edad_conductor.index[i]== 'DE 21 A 24 AÑOS') | \
                (edad_conductor.index[i]== 'DE 25 A 29 AÑOS') | (edad_conductor.index[i]== 'DE 30 A 34 AÑOS')):
                
                n_cond_18 = n_cond_18 + edad_conductor.values[i]
                
            elif ((edad_conductor.index[i]== 'DE 35 A 39 AÑOS') | (edad_conductor.index[i]== 'DE 40 A 44 AÑOS') | \
                (edad_conductor.index[i]== 'DE 45 A 49 AÑOS')):
                                
                n_cond_35 = n_cond_35 + edad_conductor.values[i]
                
            elif ((edad_conductor.index[i]== 'DE 50 A 54 AÑOS') | (edad_conductor.index[i]== 'DE 55 A 59 AÑOS') | \
                 (edad_conductor.index[i]== 'DE 60 A 64 AÑOS') | (edad_conductor.index[i]== 'DE 65 A 69 AÑOS') | \
                 (edad_conductor.index[i]== 'DE 70 A 74 AÑOS') | (edad_conductor.index[i]== 'DE MAS DE 74 AÑOS')):
                                
                n_cond_50 = n_cond_50 + edad_conductor.values[i]
       
            elif (edad_conductor.index[i]== 'DESCONOCIDA'):
                
                n_cond_desconoc = edad_conductor.values[i]               
            
    

    n_hombres = n_mujeres = n_sex_na = 0
    sexo = df_in.SEXO.value_counts()
    N = len(sexo)
    for i in range(N):
        if sexo.index[i]=="HOMBRE":
            n_hombres = sexo.values[i]
        elif sexo.index[i]=="MUJER":
            n_mujeres = sexo.values[i]
        else:
            n_sex_na = sexo.values[i]
    
    
    
    n_ilesos = n_hgraves = n_hleves = n_muertos = n_lesiv_na = 0
    lesividad = df_in.LESIVIDAD.value_counts()
    N = len(lesividad)
    for i in range(N):
        if lesividad.index[i]=="IL":
            n_ilesos = lesividad.values[i]
        elif lesividad.index[i]=="HG":
            n_hgraves = lesividad.values[i]
        elif lesividad.index[i]=="HL":
            n_hleves = lesividad.values[i]
        elif lesividad.index[i]=="MT":
            n_muertos = lesividad.values[i]
        else:
            n_lesiv_na = lesividad.values[i]

    
    
    df_out = pd.DataFrame({'num_parte': num_parte, 'num_vehic_implic': num_veh, \
                           'n_hombres': n_hombres, 'n_mujeres': n_mujeres, \
                           'n_sex_na': n_sex_na, 'n_ilesos': n_ilesos, 'n_hgraves': n_hgraves,  \
                           'n_hleves': n_hleves, 'n_muertos': n_muertos, 'n_lesiv_na': n_lesiv_na, \
                           'conductor_menor': n_cond_menor, 'conductor_18_34': n_cond_18, 'conductor_35_49': n_cond_35, \
                           'conductor_50_mas': n_cond_50, 'conductor_edad_desconoc': n_cond_desconoc}, index=[0])
    
    
    return df_out
```

## APARTE: NUEVO POSIBLE TARGET EN FUNCIÓN GRAVEDAD HERIDOS.

### Justificación umbrales función que obtiene target GRAVEDAD: 
La mayoria de accidentes son de una victima y de esos la gran mayoria de las víctimas son leves. De nuevo es un escenario muy desequilibrado

```python
dfm_accidentes.groupby('NUM_VICTIMAS')[['n_hleves','n_hgraves','n_muertos']].sum()
```

### FUNCIÓN PARA CREAR COLUMNA TARGET A PARTIR DE LA GRAVEDAD DE LOS HERIDOS.

```python
def target_gravedad(df_in, binaria=False):
    
    target = pd.Series(np.zeros(len(df_in)),dtype=np.int64)
    
    # GRAVE
    idx_muertos = set(df_in[df_in['n_muertos'] != 0].index.values)
    idx_graves = set(df_in[(df_in['n_hgraves'] != 0)].index.values)
    idx_GRAVE = idx_graves | idx_muertos
    
    # MODERADA
    idx_leves_gte_2 = set(df_in[df_in['n_hleves'] >= 2].index.values)
    
    idx_MODERADA = idx_leves_gte_2 - idx_GRAVE
    
    
    # LEVE
    idx_LEVE = (set(range(len(df_in))) - idx_GRAVE) - idx_MODERADA
    
    
    # asignaciones target
    target[idx_LEVE] = 'LEVE'
    
    if binaria == False:
        target[idx_MODERADA] = 'MODERADA'
        target[idx_GRAVE] = 'GRAVE'
    else:
        target[idx_MODERADA | idx_GRAVE] = 'MODERADA/GRAVE'
    
    return target
    
    
```

#### Aplicamos la función al subdataframe formado por las columnas relativas a gravedad del herido. Observamos reparto con 2 y 3 niveles.

```python
gravedad_accid3 = target_gravedad(dfm_accidentes[['n_hgraves', 'n_hleves', 'n_muertos']])

print('Porcentaje:\n',gravedad_accid3.value_counts(normalize=True))
print('Valores:\n',gravedad_accid3.value_counts())
```

```python
gravedad_accid2 = target_gravedad(dfm_accidentes[['n_hgraves', 'n_hleves', 'n_muertos']],binaria=True)

print('Porcentaje:\n',gravedad_accid2.value_counts(normalize=True))
print('Valores:\n',gravedad_accid2.value_counts())
```

**NOTA**: debido al bajo número y porcentaje del tercel nivel en el caso de elegir 3, y el escenario desequilibrado ya comentado, optamos por tomar dos niveles.
